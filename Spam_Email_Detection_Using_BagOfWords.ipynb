```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("movies_sentiment_data.csv")
```

```python
dataset.head()
```

```python
dataset.sentiment.value_counts()
```

```python
dataset.shape
```

```python
dataset["score"]=dataset['sentiment'].apply(lambda x:1 if x=="positive" else 0)
```

```python
dataset.head()
```

```python
dataset.score.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(dataset["review"],dataset["score"],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train[:4]
```

```python
type(x_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector=CountVectorizer()
```

```python
x_train_v=vector.fit_transform(x_train.values)
```

```python
x_train_v.toarray()
```

```python
# vector.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier=MultinomialNB()
```

```python
classifier.fit(x_train_v,y_train)
```

```python
x_test_v=vector.transform(x_test)
```

```python
predicted=classifier.predict(x_test_v)
```

```python
predicted
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predicted))
```

```python
#using Pipeline
```

```python
from sklearn.pipeline import Pipeline
```

```python
classifier=Pipeline([("vectorizer",CountVectorizer()),
                     ("nb",MultinomialNB())])
```

```python
print(classifier)
```

```python
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
print(classification_report(y_test,y_predict))
```
